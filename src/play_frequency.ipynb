## 파일 불러오기

```python
import pandas as pd
import seaborn as sns; sns.set()
```

```python
df = pd.read_csv('빅데이터 분석 공모전 raw data 추출_2020.12.16-2020.12.31.csv')
```

## 연령 40~60대 사이로 줄이기

```python
df = df[(df['연령']>1961) & (df['연령']<= 1981)]
```

### 네이버 예약만

```python
df = df[df['전송사업자명']=='네이버예약']
```

```python
df.columns
```

## 제작사별로 count

```python
count={}
for i in df['기획제작사명']:
    try: count[i] += 1
    except: count[i]=1
```

### 내림차순으로 정렬
* 과정:
* 1) 약 상위 10개 정도의 기획사를 확인하여,
* 2) 추출했던 csv 파일에서 해당 기획사 공연의 출연진이 누구인지 확인한다.
* 3) 출연진 검색을 통해, 해당 공연명을 확인한다.
